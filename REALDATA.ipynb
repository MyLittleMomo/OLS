```python
from sklearn import random
import numpy as np
import math
import numpy as np
from sklearn import linear_model
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_diabetes
from sklearn import linear_model
diabetes = load_diabetes() #10 features 442 datapoints
m=400 #you can vary this parameter between [10-400], but I varied between [20-400] by adding 10 because error bound formula assumes that there are more datapoints than features
d=10 #you can vary this parameter between [1-10]
delta=0.01 #fixed small delta in order for error bound formula work
x_test=diabetes.data[-42:,-d:] #I have used last 42 datapoints as test sample with d features (starting from last feature)
y_test=diabetes.target[-42:]
x_train=diabetes.data[:m,-d:]
y_train=diabetes.target[:m]
model = linear_model.LinearRegression(fit_intercept=False)
model.fit(x_train, y_train)
print('Coefs: \n', model.coef_)
y_pred =model.predict(x_test)
print("Mean squared error: %.2f" % mean_squared_error(y_test,y_pred))
RH = mean_squared_error(y_test, y_pred)
print("Real error RH = ",RH)
w_star=model.coef_
#print(w_star)
L=[None]*(len(x_train));
for i in range(0,len(x_train)):
  L[i]=(abs(np.dot(w_star, x_train[i])-y_train[i]))*(abs(np.dot(w_star, x_train[i])-y_train[i]))
  #print("Error on training point ",i+1," = ",L[i])
M=max(L);
print('Loss function bound M =',M)
LH=0;
for j in range(0,len(x_train)):
  LH=LH+L[j];
RS=abs(LH/len(x_train));
print('Empirical risk RS = ',RS);
ErrorBound=RS+M*math.sqrt((2*d*math.log(math.e*m/d))/m) +M*math.sqrt((math.log(1/delta))/(2*m))
print('Error bound = ',ErrorBound);
TEST=ErrorBound-RH;
print("Difference between Error bound and Real error = ",TEST);

```
